##5.2 Validation Using Mere Pandas

```python
df.isna().sum()
```

```python
rules = {
    "credit_score": (300, 850),
    "years_employed": (0, None),
    "annual_income": (0, None),
    "loan_amount": (0, None),
    "loan_to_income_ratio": (0, None),
    "payment_to_income_ratio": (0, 1)
}

result = True

for col in rules:

    min_val, max_val = rules[col]

    print(f"\nValidasi kolom: {col}")

    # cek numeric
    type_check = pd.api.types.is_numeric_dtype(df[col])
    print("Numeric:", type_check)

    # cek min
    min_check = (df[col] >= min_val).all()
    print("Min valid:", min_check)

    # cek max jika ada
    if max_val is not None:
        max_check = (df[col] <= max_val).all()
        print("Max valid:", max_check)
        result = result & type_check & min_check & max_check
    else:
        result = result & type_check & min_check

print("\nHASIL VALIDASI KESELURUHAN:", result)
```

#6. Formatting Errors

```python
def convert_years(x):
    if pd.isna(x):
        return x

    parts = str(x).split(".")

    if len(parts) == 2:
        years = int(parts[0])
        months = float(parts[1])
        return years + round(months/12, 2)
    else:
        return float(x)

df['years_employed'] = df['years_employed'].apply(convert_years)
```

```python
column_ = "credit_history_years"
mean_, std_, median_, q1_, q3_ = df[column_].describe()[['mean','std','50%','25%','75%']].values
mean_, std_, median_, q1_, q3_

def is_outlier_zscore(val_, mean_=mean_, std_=std_):
  return (val_ - mean_)/std_ > 3 or (val_ - mean_)/std_ < -3

def is_outlier_iqr(val_, q1_=q1_, q3_=q3_):
  iqr_ = q3_ - q1_
  return val_ > q3_ + 1.5*iqr_ or val_ < q1_ - 1.5*iqr_

df['is_outliers'] = df[column_].apply(lambda x: is_outlier_zscore(x))
# df['is_outliers'] = df[column_].apply(lambda x: is_outlier_iqr(x))
```

#8. Feature Engineering (Create New Column - Savings_Assets_Ratio)

##8.1 Scaling

```python
#Feature Engineering

df['savings_assets_ratio'] = df.apply(
    lambda x: x['savings_assets'] / x['annual_income']
    if x['annual_income'] != 0 else 0,
    axis=1
)

# Min-Max Scaling

scale_cols = [
    'annual_income',
    'loan_amount',
    'years_employed',
    'loan_to_income_ratio',
    'payment_to_income_ratio',
    'debt_to_income_ratio',
    'interest_rate',
    'savings_assets_ratio'
]

for col in scale_cols:
    min_ = df[col].min()
    max_ = df[col].max()
    df[col] = (df[col] - min_) / (max_ - min_)


# Verifikasi

df[scale_cols].describe()
```

#10. Uji dengan Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# prepare for splitting
target = df['loan_status'].values
features = df.drop(columns=["loan_status", "is_outliers"]).values
```

```python
# splitting
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)
print(len(X_train), len(X_test), len(y_train), len(y_test))
print()

# training
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
```

```python
# predict
y_pred = dtc.predict(X_test)

# evaluation
print(classification_report(y_pred, y_test))
```
